```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 데이터 로드 
df = pd.read_csv('./data/melb_data.csv')
df
```

```python
# 데이터 구조 및 기초 통계 확인
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# 결측치 탐색 
df.isnull().sum()
```

```python
# 결측치 확인 및 처리 
'''Address(주소), SellerG(판매자 그룹), Date(거래 날짜), BuildingArea(건물 면적), 
YearBuilt(건축연도)'''
df = df.dropna()

drop_columns = ["Address", "SellerG", "Date", "BuildingArea", "YearBuilt"]
df = df.drop(columns=drop_columns)
```

```python
# 범주형 변수 인코딩 
df = pd.get_dummies(df, drop_first=True)
```

```python
# 상관관계 분석 및 시각화 
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap="viridis", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# 가격과 주요 변수 간 관계 시각화
sns.pairplot(df, vars=["Price", "Rooms", "Distance", "Landsize"], diag_kind="kde")
plt.show()
```

```python
# 인사이트 도출
# 1. 방(Rooms)이 많을수록 가격(Price)이 높아지는 경향
# 2. 도심에서 멀어질수록(Distance) 가격이 낮아짐
# 3. 토지 크기(Landsize)와 가격 간의 상관관계는 약함
```
