```python
def outlier_iqr(data, column): 

    global lower, upper    
 
    q1, q3 = np.quantile(data[column], 0.25), np.quantile(data[column], 0.75)          
  
    iqr = q3 - q1 
 
    cut_off = iqr * 1.5          
   
    lower, upper = q1 - cut_off, q3 + cut_off     
    
    print('IQR: ',iqr)     
    print('lower bound: ', lower)     
    print('upper bound: ', upper)    
    
    outlier_upper = data[data[column] > upper]     
    ouliter_lower = data[data[column] < lower]    
 
    return print('Number of outlier in column [' + column + ']:', outlier_upper.shape[0] + ouliter_lower.shape[0], end='\n\n')
```

```python
df_v.columns

for i in df_v.columns:
    outlier_iqr(df_v, i)
```

```python
import matplotlib.pyplot as plt

filtered_vars = df_corr[abs(df_corr['Price']) > 0.3].index
filtered_vars = filtered_vars.drop('Price')

plt.figure(figsize=(12, len(filtered_vars) * 3))

for i, var in enumerate(filtered_vars, 1):
    plt.subplot(len(filtered_vars), 2, i)
    plt.scatter(df_v[var], df_v['Price'], alpha=0.5)
    plt.xlabel(var)
    plt.ylabel('Price')
    plt.title(f'Price vs {var}')

plt.tight_layout()
plt.show()
```
